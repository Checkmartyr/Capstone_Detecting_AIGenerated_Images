```python

feature_names = [col for col in data.columns if col != 'Class']

# Determine the number of rows and columns for the subplots grid
num_features = len(feature_names)
num_cols = 4  # Number of columns in the grid
num_rows = (num_features // num_cols) + (num_features % num_cols > 0)

# Create a figure and axes
fig, axes = plt.subplots(num_rows, num_cols, figsize=(20, num_rows * 5))
axes = axes.flatten()

# Create KDE plots for all features
for i, feature_name in enumerate(feature_names):
    sns.kdeplot(data=data, x=feature_name, hue='Class', fill=True, common_norm=False, multiple="stack", alpha=.5, linewidth=1, ax=axes[i])
    axes[i].set_title(f'Distribution of {feature_name} by Class')
    axes[i].set_xlabel(feature_name)
    axes[i].set_ylabel('Density')

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```

```python


feature_names = [col for col in data.columns if col != 'Class']

# Define the number of columns per figure
columns_per_figure = 4

# Loop through the features in chunks and create separate figures
for i in range(0, len(feature_names), columns_per_figure):
    # Determine the current set of features to plot
    current_features = feature_names[i:i + columns_per_figure]
    
    # Determine the number of columns needed for the current set of features
    num_cols = len(current_features)
    
    # Create a figure and axes
    fig, axes = plt.subplots(1, num_cols, figsize=(num_cols * 5, 5))
    
    # Ensure axes is always a list
    if num_cols == 1:
        axes = [axes]
    
    # Plot each feature on a separate subplot
    for j, feature_name in enumerate(current_features):
        sns.kdeplot(data=data, x=feature_name, hue='Class', fill=True, common_norm=False, multiple="stack", alpha=.5, linewidth=.5, ax=axes[j])
        axes[j].set_title(f'Distribution of {feature_name} by Class')
        axes[j].set_xlabel(feature_name)
        axes[j].set_ylabel('Density')

    # Adjust layout
    plt.tight_layout()
    
    # Save the figure
    plt.savefig(f'./img/feature_distribution_{i//columns_per_figure + 1}.png')
    
    # Close the figure to free up memory
    plt.close(fig)
```

```python
from PIL import Image
import os

# Directory containing your images
image_directory = 'img'

# List of image file names (ensure these are the actual filenames in your directory)
image_files = os.listdir(image_directory)

# Create full paths for each image
image_paths = [os.path.join(image_directory, img) for img in image_files]

# Open all images and store them in a list
images = [Image.open(img_path) for img_path in image_paths]

# Assuming all images are the same size
square_size = 400

# Open all images, resize them to square, and store them in a list
images = []
for img_path in image_paths:
    img = Image.open(img_path)
    img = img.resize((square_size, square_size))
    images.append(img)

# Calculate the size of the final image
columns = 5  # number of columns
rows = 5  # number of rows
final_image_width = square_size * columns 
final_image_height = square_size * rows


# Create a new blank image with the calculated size
# Create a new blank image with the calculated size
final_image = Image.new('RGB', (final_image_width, final_image_height))

# Paste each image into the final image
for index, img in enumerate(images):
    x_offset = (index % columns) * square_size
    y_offset = (index // columns) * square_size
    final_image.paste(img, (x_offset, y_offset))

# Save the final image
final_image.save('merged_image.jpg')
final_image.show()

print("Images have been resized and merged successfully!")
```
